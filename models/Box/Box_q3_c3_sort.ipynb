```python
directory_path = r'D:\origin-source-code-bill\models\KMeans\cluster3'
directory_path_box = 'D:\origin-source-code-bill\models\Box'
pkl_files = [f for f in os.listdir(directory_path) if f.endswith('.pkl')]
print(pkl_files)


list = []

for pkl_file in pkl_files:
    file_path = os.path.join(directory_path, pkl_file)
    # print("file :::" , file_path)
    variable_name = os.path.splitext(pkl_file)[0]
    # print("Var ::" , variable_name)
    df_col_combined = pd.read_pickle(file_path)
    print( df_col_combined.columns)
    
    date = pd.to_timedelta(df_col_combined['total_time'])
    time_hours = date.dt.total_seconds() / 3600
    
    df_col_combined['hours'] = time_hours
    # print("DF ::" , df_col_combined.columns)
    df_sorted_hours = df_col_combined.sort_values(by='hours' , ascending=True)
    
    for index, row in sorted_df.iterrows():
        # row[0] is 'col'
        # row[1] is time1 127.343611
        # row[2] is time1  70.730972
        # row[3] is time1  35.859043`
        if df_sorted_hours.columns[-4] == row[0]:
            col = row[0]
            total = df_sorted_hours['hours'].sum() 
            
            p1 = df_sorted_hours[df_sorted_hours['hours'] <= row[3]]
            percentage_p1 = (p1['hours'].sum() / total) *100
            print( 'Pre 1 :::' , percentage_p1)
            
            p2 = df_sorted_hours[df_sorted_hours['hours'].between(row[3], row[1])]
            percentage_p2 = (p2['hours'].sum() / total) 
            print( 'Pre 2 :::' , percentage_p2)
            
            # p2 = df_sorted_hours[df_sorted_hours['hours'].between(row[3], row[2])]
            # p3 = df_sorted_hours[df_sorted_hours['hours'].between(row[2] , row[1])]
            
            p4 = df_sorted_hours[df_sorted_hours['hours'] >= row[1]]
            percentage_p3 = (p4['hours'].sum() / total)
            print( 'Pre 3 :::' , percentage_p3)
            
        # else:
            # print("sorted data hours")
    
    data = {
        'col' : col ,
        'shape_p1' : p1.shape , 
        'shape_p2' : p2.shape , 
        # 'shape_p3' : p3.shape , 
        'shape_p4' : p4.shape  ,
        'Percentage_p1' : percentage_p1,
        'Percentage_p2' : percentage_p2 ,
        'Percentage_p3' : percentage_p3 
    }

    list.append(data)
    data_df = pd.DataFrame(list)
    print(data_df)
```
